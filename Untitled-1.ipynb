```python
# Set the seed
# np.random.seed()
import numpy as np
from functions import *
m = 1000
n = 300

A = np.random.randn(m, n)
x0 = np.random.random(n)
b = A.dot(x0) + np.random.random(m)
c = np.random.randn(n)

f = lambda x: c.dot(x)-np.sum(np.log(b-A.dot(x)))
grad_f = lambda x: c+A.T.dot(1/(b-A.dot(x)))
nabla_f = lambda x: A.T.dot(np.diag(1/(b-A.dot(x))**2)).dot(A)

decrement = lambda dx, x: (dx.dot(nabla_f(x).dot(dx)))**2/2
dom_f = lambda x: np.all(b-A.dot(x)>0)
```

```python
n = A.shape[1]
x = cp.Variable(n)
objective = cp.Minimize(c.T @ x- cp.sum(cp.log(b-A @ x)))    
constraints = [A @ x <= b]

# Form and solve problem.
prob = cp.Problem(objective, constraints)
cvx_solution = prob.solve(solver='MOSEK')
```

```python
v_g = gradient_descent(f, grad_f, x0.copy(), A, b, dom_f, MAXITERS=2000, TOL=1e-8, alpha=0.001, beta=0.8, print_iter=False)
```

```python
v_n = newton(f, grad_f, nabla_f, x0.copy(), A, b, dom_f, MAXITERS=1000, TOL=1e-8, alpha=0.001, beta=0.8, print_iter=True, N=1, diag_only=False, decrement_func=decrement, eq=False)
```

```python
plot_error_iter(v_g, cvx_solution, "Gradient descent", color='r')
plot_error_iter(v_n, cvx_solution, "Newton's method")
plt.legend()
plt.show()
```
